```python
from google.colab import files
import pandas as pd
import os
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np



# Subir el archivo de Excel
uploaded = files.upload()

for filename in uploaded.keys():
    df = pd.read_excel(filename)
    print("Archivo cargado:", filename)
```

```python
# Obtiene la lista de archivos en el directorio actual
archivos = os.listdir()

# Filtra solo los archivos con extensión .xls
Oraculo_xls = [archivo for archivo in archivos if archivo.endswith('.xlsx')]

if len(Oraculo_xls) == 0:
    print("No se encontraron archivos en formato Excel (.xls) en el directorio actual.")
else:
    # Tomamos el primer archivo encontrado en la lista
    Oraculo = Oraculo_xls[0]

    # Lee el archivo Excel y carga los datos en un DataFrame de pandas
    Oraculo = pd.read_excel(Oraculo)
    Oraculo.set_index('FECHA', inplace=True)
```

```python
# Supongamos que 'Oraculo' es tu DataFrame y que ya tiene un índice de tipo datetime
Oraculo.index = pd.to_datetime(Oraculo.index)

# Agrupar columnas según el sufijo
TES_columns = [col for col in Oraculo.columns if col.endswith('TES')]
FS_columns = [col for col in Oraculo.columns if col.endswith('FS')]
IBR_columns = [col for col in Oraculo.columns if col.endswith('IBR')]
UVR_columns = [col for col in Oraculo.columns if col.endswith('UVR')]
IPC_columns = [col for col in Oraculo.columns if col.endswith('IPC')]
REPO_columns = [col for col in Oraculo.columns if 'REPO' in col]

# Crear un diccionario con las agrupaciones
column_groups = {
    'TES': TES_columns,
    'FS': FS_columns,
    'IBR': IBR_columns,
    'UVR': UVR_columns,
    'IPC': IPC_columns,
    'REPO': REPO_columns
}

# Crear listas desplegables con las agrupaciones
dropdown_group = widgets.Dropdown(
    options=list(column_groups.keys()),
    value='IPC',
    description='Grupo:',
)

# Inicializar las opciones de columnas basadas en el grupo seleccionado
dropdown1 = widgets.Dropdown(options=IPC_columns, value=IPC_columns[0], description='Opción 1:')
dropdown2 = widgets.Dropdown(options=IPC_columns, value=IPC_columns[1], description='Opción 2:')

# Crear un selector de año y mes de inicio
year_month_start = widgets.SelectionSlider(
    options=[f"{date.year}-{date.month:02d}" for date in Oraculo.index],
    value=f"{Oraculo.index.min().year}-{Oraculo.index.min().month:02d}",
    description='Fecha inicio (Año-Mes):',
    continuous_update=False,
)

# Crear un selector de año y mes de fin
year_month_end = widgets.SelectionSlider(
    options=[f"{date.year}-{date.month:02d}" for date in Oraculo.index],
    value=f"{Oraculo.index.max().year}-{Oraculo.index.max().month:02d}",
    description='Fecha fin (Año-Mes):',
    continuous_update=False,
)

# Función para actualizar las opciones disponibles según el grupo seleccionado
def actualizar_opciones(grupo):
    opciones = column_groups[grupo]
    dropdown1.options = opciones
    dropdown2.options = opciones
    dropdown1.value = opciones[0]
    dropdown2.value = opciones[1]

# Función para actualizar la gráfica y mostrar los cálculos estadísticos
def actualizar_grafico(columna1, columna2, fecha_inicio, fecha_fin):
    # Convertir las selecciones a datetime
    fecha_inicio = pd.to_datetime(f"{fecha_inicio}-01")  # Agregar un día para convertir a datetime
    fecha_fin = pd.to_datetime(f"{fecha_fin}-01") + pd.offsets.MonthEnd(1)  # Fin del mes

    # Filtrar el DataFrame por el rango de fechas seleccionado
    data_rango = Oraculo.loc[fecha_inicio:fecha_fin]

    # Realizar la operación (columna1 - columna2) * 100
    operacion = (data_rango[columna1] - data_rango[columna2]) * 100

    # Calcular estadísticas
    ultimo_valor = operacion.iloc[-1]
    promedio = operacion.mean()
    desviacion = operacion.std()
    mediana = operacion.median()
    maximo = operacion.max()
    minimo = operacion.min()

    # Crear una figura con tres gráficos: columnas individuales, operación y el histograma
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6), gridspec_kw={'width_ratios': [2, 2, 1]})

    # Gráfico 2: Mostrar ambas columnas (columna1 y columna2) en la primera posición
    ax1.plot(data_rango.index, data_rango[columna1], label=columna1, color='blue')
    ax1.plot(data_rango.index, data_rango[columna2], label=columna2, color='green')
    ax1.set_title(f'Gráfico de {columna1} y {columna2}')
    ax1.set_xlabel('Fecha (Mes-Año)')
    ax1.set_ylabel('Valor')

    # Formato de fecha en el eje X
    ax1.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m'))
    fig.autofmt_xdate()  # Rotar las etiquetas de la fecha para mejor visualización

    ax1.legend()

    # Gráfico principal: operación (columna1 - columna2) * 100 en la segunda posición
    ax2.plot(data_rango.index, operacion, label=f'({columna1} - {columna2}) * 100', color='black')
    ax2.set_title(f'Gráfico de ({columna1} - {columna2}) * 100')
    ax2.set_xlabel('Fecha (Mes-Año)')
    ax2.set_ylabel('Resultado de la operación')

    # Agregar líneas de promedio, máximo, mínimo y desviaciones estándar
    ax2.axhline(promedio, color='blue', linestyle='--', label=f'Promedio: {promedio:.2f}')
    ax2.axhline(maximo, color='green', linestyle='--', label=f'Máximo: {maximo:.2f}')
    ax2.axhline(minimo, color='red', linestyle='--', label=f'Mínimo: {minimo:.2f}')
    ax2.axhline(promedio + desviacion, color='purple', linestyle=':', label=f'1 Desv: {desviacion:.2f}')
    ax2.axhline(promedio - desviacion, color='purple', linestyle=':')

    # Formato de fecha en el eje X
    ax2.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m'))
    fig.autofmt_xdate()  # Rotar las etiquetas de la fecha para mejor visualización

    ax2.legend()

    # Gráfico de frecuencias: histograma del resultado de la operación en la tercera posición
    ax3.hist(operacion, bins=20, orientation='horizontal', color='red', edgecolor='black')
    ax3.set_title('Frecuencia de la Operación')
    ax3.set_xlabel('Frecuencia')
    ax3.set_ylabel('Valor')

    # Mostrar etiquetas fuera del gráfico
    ax3.yaxis.tick_right()
    ax3.yaxis.set_label_position("right")

    # Mostrar estadísticas en el gráfico de la derecha
    stats_text = f'''
    Último valor: {ultimo_valor:.2f}
    Promedio: {promedio:.2f}
    Desviación estándar: {desviacion:.2f}
    Mediana: {mediana:.2f}
    Máximo: {maximo:.2f}
    Mínimo: {minimo:.2f}
    '''
    ax3.text(0.95, 0.95, stats_text, transform=ax3.transAxes, fontsize=10,
             verticalalignment='top', horizontalalignment='right', bbox=dict(facecolor='white', alpha=0.6))

    plt.tight_layout()
    plt.show()

# Conectar widgets a las funciones
dropdown_group.observe(lambda change: actualizar_opciones(change['new']), names='value')
out = widgets.interactive_output(actualizar_grafico, {
    'columna1': dropdown1,
    'columna2': dropdown2,
    'fecha_inicio': year_month_start,
    'fecha_fin': year_month_end
})

# Mostrar los widgets y el gráfico
display(dropdown_group, dropdown1, dropdown2, year_month_start, year_month_end, out)
```
