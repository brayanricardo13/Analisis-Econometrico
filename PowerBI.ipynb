```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Crear un DataFrame con datos ficticios
data = {
    'Fecha': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'Ventas': np.random.randint(50, 200, size=10),
    'Ganancias': np.random.uniform(100, 500, size=10)
}

df = pd.DataFrame(data)

# Mostrar el DataFrame
print("DataFrame:")
print(df)

# Gráfico de líneas para las ventas
plt.figure(figsize=(10, 5))
plt.plot(df['Fecha'], df['Ventas'], marker='o', linestyle='-')
plt.title('Ventas a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()

# Gráfico de barras para las ganancias
plt.figure(figsize=(10, 5))
plt.bar(df['Fecha'], df['Ganancias'], color='green', alpha=0.7)
plt.title('Ganancias a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Ganancias')
plt.grid(axis='y')
plt.show()

# Diagrama de dispersión entre Ventas y Ganancias
plt.figure(figsize=(8, 8))
plt.scatter(df['Ventas'], df['Ganancias'], color='red', marker='o')
plt.title('Relación entre Ventas y Ganancias')
plt.xlabel('Ventas')
plt.ylabel('Ganancias')
plt.grid(True)
plt.show()
```

```
auth = DeviceCodeLoginAuthentication()
```

```
PBI_Visuallize= quick_visualize.QuickVisualize(get_dataset_config(df),auth = auth)
```
