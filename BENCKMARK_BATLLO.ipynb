```python
import pandas as pd
import matplotlib.pyplot as plt

# Definir la variable inicio_fila
inicio_fila = 6  # Fila desde la que comenzar a leer los datos

# Leer los datos desde el archivo Excel
data_colcap = pd.read_excel("/content/BMK BAT.xlsx", sheet_name="Vlr Unidad", header=None, skiprows=inicio_fila-1, usecols="B,G", names=["Fechas", "Valores_Colcap"])
data_coltes = pd.read_excel("/content/BMK BAT.xlsx", sheet_name="Vlr Unidad", header=None, skiprows=inicio_fila-1, usecols="B,F", names=["Fechas", "Valores_Coltes"])
data_fic_bbva = pd.read_excel("/content/BMK BAT.xlsx", sheet_name="Vlr Unidad", header=None, skiprows=inicio_fila-1, usecols="B,D", names=["Fechas", "Valores"])
data_multiescala = pd.read_excel("/content/BMK BAT.xlsx", sheet_name="Vlr Unidad", header=None, skiprows=inicio_fila-1, usecols="B:C", names=["Fechas", "Valores"])

# Calcular los ndices base 1000
index_colcap = (data_colcap["Valores_Colcap"] / data_colcap["Valores_Colcap"].iloc[0]) * 1000
index_coltes = (data_coltes["Valores_Coltes"] / data_coltes["Valores_Coltes"].iloc[0]) * 1000
index_fic_bbva = (data_fic_bbva["Valores"] / data_fic_bbva["Valores"].iloc[0]) * 1000
index_multiescala = (data_multiescala["Valores"] / data_multiescala["Valores"].iloc[0]) * 1000

# Crear el grfico combinado
plt.figure(figsize=(10, 6))

# Grfico 1 - Comparacin de ndices Base 1000 entre COLCAP y COLTES CP BVC
plt.plot(data_colcap["Fechas"], index_colcap, label="COLCAP Index 1000")
plt.plot(data_coltes["Fechas"], index_coltes, label="COLTES CP BVC Index 1000")

# Grfico 2 - FIC plazo 30 BBVA
plt.plot(data_fic_bbva["Fechas"], index_fic_bbva, label="FIC plazo 30 BBVA Index 1000")

# Grfico 3 - Multiescala davivienda
plt.plot(data_multiescala["Fechas"], index_multiescala, label="Multiescala davivienda Index 1000")

plt.title("Comparacin de ndices Base 1000")
plt.xlabel("Fechas")
plt.ylabel("ndice (Base 1000)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
